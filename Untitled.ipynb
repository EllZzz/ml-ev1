```python
# 🔹 IMPORTS
import pandas as pd

# 🔹 CARGAR DATASETS LIMPIOS DESDE CSVs (generados por el pipeline)
cardio_clean = pd.read_csv("data/02_intermediate/cardio_data_processed_clean.csv")
lifestyle_clean = pd.read_csv("data/02_intermediate/smoking_drinking_clean.csv")
heart_clean = pd.read_csv("data/02_intermediate/heart_2020_cleaned_clean.csv")

# 🔹 RESÚMENES DE LIMPIEZA
def resumen_limpieza(df, nombre):
    filas_originales = len(df) + df.duplicated().sum()  # estimación aproximada
    duplicados_eliminados = df.duplicated().sum()
    nulos_totales = df.isnull().sum().sum()
    resumen = pd.DataFrame({
        "dataset": [nombre],
        "filas_originales": [filas_originales],
        "duplicados_eliminados": [duplicados_eliminados],
        "nulos_totales": [nulos_totales]
    })
    return resumen

print("==== CARDIO SUMMARY ====")
print(resumen_limpieza(cardio_clean, "cardio"))

print("\n==== LIFESTYLE SUMMARY ====")
print(resumen_limpieza(lifestyle_clean, "lifestyle"))

print("\n==== HEART SUMMARY ====")
print(resumen_limpieza(heart_clean, "heart_data"))

# 🔹 ESTADÍSTICAS DE BMI EN CARDIO
print("\n==== Estadísticas de BMI en Cardio ====")
print(cardio_clean["bmi"].describe())

# 🔹 DETECCIÓN DE OUTLIERS DE BMI
Q1 = cardio_clean["bmi"].quantile(0.25)
Q3 = cardio_clean["bmi"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_mask = (cardio_clean["bmi"] < lower_bound) | (cardio_clean["bmi"] > upper_bound)
outliers = cardio_clean[outliers_mask]

print(f"Cantidad de outliers de BMI detectados: {len(outliers)}")
if len(outliers) > 0:
    print("\nPrimeros outliers de BMI detectados:")
    display(outliers.head())
```

```python
import pandas as pd

# 🔹 Cargar dataset limpio
df = pd.read_csv("data/02_intermediate/cardio_data_processed_clean.csv")

# 🔹 Crear bmi2 recalculado a partir de peso y altura
df["bmi2"] = df["weight"] / ((df["height"]/100) ** 2)

# 🔹 Ver estadísticas
print("==== Estadísticas de BMI2 en Cardio ====\n")
print(df["bmi2"].describe())

# 🔹 Detectar outliers usando IQR
Q1 = df["bmi2"].quantile(0.25)
Q3 = df["bmi2"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["bmi2"] < lower_bound) | (df["bmi2"] > upper_bound)]

print(f"\nCantidad de outliers de BMI2 detectados: {outliers.shape[0]}")
print("\nPrimeros outliers de BMI2 detectados:")
display(outliers.head())
```
